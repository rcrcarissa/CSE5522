# **Problem 2: Logistic Regression Problem**

2.0: Set up the environment.

```
import pandas as pd
import numpy as np
```

2.1: Read csv values; initialize weights and learning rate.

```
IcecreamUrl = "https://raw.githubusercontent.com/efosler/cse5522data/master/icecream_logisticregression.csv"
icecream_dataframe = pd.read_csv(IcecreamUrl)
icecream_dataframe.head()
temperature = np.array(icecream_dataframe['Temperature'],dtype='float')
icecream = np.array(icecream_dataframe['IceCream'],dtype='float')

weights = np.zeros(2)
laerning_rate = 0.001
```

2.2: Define functions to compute loss, to compute gradient, and to update weights.

```
def computeMSELoss(w, x, y):
  loss = np.power(y - np.power(1 + np.exp(w[1] * x + w[0]), -1), 2)
  loss = np.average(loss)
  return loss

def computeMSEBatchGradient(w, x, y):
  w0 = w[0]
  w1 = w[1]
  d0 = 2 * (y - np.power(1 + np.exp(w1 * x + w0), -1)) * np.power(1 + np.exp(w1 * x + w0), -2) * np.exp(w1 * x + w0)
  d1 = x * d0
  d0 = np.average(d0)
  d1 = np.average(d1)
  return np.array([d0, d1])

def computeGradientDescentWeightUpdate(w, x, y, lr):
  d = computeMSEBatchGradient(w, x, y)
  w -= lr * d
  return w
```

2.3: Gradient descent process with No-Improvement-In-10 early stop strategy.

```
epochs = 0
previous_loss = np.inf
epochs_no_improvement = 0
while True:
  loss = computeMSELoss(weights, temperature, icecream)

  # terminate condition
  if loss < previous_loss:
    epochs_no_improvement = 0
  else:
    epochs_no_improvement += 1
    if epochs_no_improvement >= 10:
      break
  
  weights = computeGradientDescentWeightUpdate(weights, temperature, icecream, laerning_rate)
  previous_loss = loss
  epochs += 1

  ############# track loss and weights #############
  if epochs % 100000 == 0:
    print("Epoch "+str(epochs)+ " has done!")
    print(loss)
    print(weights)
print(epochs)
print(loss)
print(weights)
```
